```python
def est_poission_rate_ci(n_events, 
                         source_duration, 
                         target_duration=1.0, 
                         ci=0.98,
                         expected_avg_n_events_per_unit_duration=200):
    """
    Estimate poission rate parameter confidence interval

    Parameters
    ----------
    n_events : int
        number of observed events
    source_duration : float
        duration for which n_events were observed
    target_duration : float
        duration for which to return expected rate
    ci : float
        size of confidence interval

    Returns
    -------
    (float,float)
        Confidence interval left, right bounds
    """
    
    def calc_posterior(prior_shape, prior_scale, n_events, duration):
        posterior_shape = prior_shape + n_events
        posterior_scale = prior_scale/(prior_scale*duration+1)

        return posterior_shape, posterior_scale

    rate_prior_average = expected_avg_n_events_per_unit_duration
    rate_prior_shape = 1
    rate_prior_scale = rate_prior_average/rate_prior_shape
    rate_prior = stats.gamma(rate_prior_shape,scale=rate_prior_scale)
    
    posterior_shape, posterior_scale = calc_posterior(rate_prior_shape, 
                                                      rate_prior_scale, 
                                                      n_events, 
                                                      source_duration/target_duration)
    posterior = stats.gamma(posterior_shape,scale=posterior_scale)
    return posterior
```

```python
def fetch_grouped_data(dimension, values=[]):

    project_id = 'distribusion-next-13'
    
    filter_line = '' if not len(values) else f'and {dimension} in ("' + '", "'.join([v for v in values]) + '")'

    sql = f'''
    with tableA as (
      select *
      from
        (
          SELECT date_add(dt, interval -60 minute) as booking_date_time_from, 
           dt as booking_date_time_to,
          FROM unnest (GENERATE_TIMESTAMP_ARRAY( date_add(timestamp_trunc(current_timestamp, minute ), interval -2 day),
                                    current_timestamp, INTERVAL 1 hour)) as dt
        ) t1
        cross join
        (
          select distinct {dimension} as a_d_dim_src
          from `distribusion-next-13.dl_analytics_v_0_2.bookings_caramelo` tt
          where date(tt.booked_at) >= date_add(current_date(), interval -2 day)
          {filter_line}
        )
      ),

      tableB as ( 
      select {dimension} as a_d_dim,
             t.booked_at as booked_at,
             t.tickets_pax
      FROM `distribusion-next-13.dl_analytics_v_0_2.bookings_caramelo` t
      Where date(t.booked_at) >= date_add(current_date(), interval -2 day)
            #and t.booked_at < date_add(current_date(), interval -48 hour)
      )

    SELECT a.*, 
          coalesce(sum(b.tickets_pax),0) as pax
    FROM tableA a
    LEFT JOIN tableB b
    ON timestamp(a.booking_date_time_from) <= timestamp(b.booked_at)
        and timestamp(a.booking_date_time_to) > timestamp(b.booked_at)
        and a.a_d_dim_src = b.a_d_dim
    group by booking_date_time_from, booking_date_time_to, a_d_dim_src

  
    '''
    
    df = pandas_gbq.read_gbq(sql, project_id=project_id)

    return df
```

```python
def get_bigq(dimension,timestamp,values=[]):

    project_id = 'distribusion-next-13'
    
    filter_line = '' if not len(values) else f'and {dimension} in ("' + '", "'.join([v for v in values]) + '")'

    sql = f'''
    with tableA as (
      select *
      from
        (
          SELECT date_add(dt, interval -60 minute) as booking_date_time_from, 
           dt as booking_date_time_to,
          FROM unnest (GENERATE_TIMESTAMP_ARRAY( date_add(TIMESTAMP({timestamp}), interval -2 day),
                                    TIMESTAMP({timestamp}), INTERVAL 1 hour)) as dt
        ) t1
        cross join
        (
          select distinct {dimension} as a_d_dim_src
          from `distribusion-next-13.dl_analytics_v_0_2.bookings_caramelo` tt
          where date(tt.booked_at) >= date_add(date(TIMESTAMP({timestamp})), interval -2 day)
          {filter_line}
        )
      ),

      tableB as ( 
      select {dimension} as a_d_dim,
             t.booked_at as booked_at,
             t.tickets_pax
      FROM `distribusion-next-13.dl_analytics_v_0_2.bookings_caramelo` t
      Where date(t.booked_at) >= date_add(date(TIMESTAMP({timestamp})), interval -2 day)
      )

    SELECT a.*, 
          coalesce(sum(b.tickets_pax),0) as pax
    FROM tableA a
    LEFT JOIN tableB b
    ON timestamp(a.booking_date_time_from) <= timestamp(b.booked_at)
        and timestamp(a.booking_date_time_to) > timestamp(b.booked_at)
        and a.a_d_dim_src = b.a_d_dim
    group by booking_date_time_from, booking_date_time_to, a_d_dim_src

  
    '''
    df = pandas_gbq.read_gbq(sql, project_id=project_id)

    return df
```

```python
def add_constant(matrix):
    matrix[:, 3] = matrix[:, 3] + 1
    return matrix

def sub_constant(matrix):
    matrix[:, 3:5] = matrix[:, 3:5] - 1   
    return matrix

def filter_matrix(matrix, id):
    return matrix[matrix[:, 2] == id]

def get_unique_vals(matrix):
    return np.unique(matrix[: , 2])

def extract_partner_data(matrix):
    partner_data = []
    partner_id = get_unique_vals(matrix)
    for id in partner_id:
        sub_matrix = filter_matrix(matrix, id)
        partner_data.append(sub_matrix)
    return partner_data
```

```python
def get_lower_ci(n_obs_events, tolerance_threshold):
    samples = sliding_window_view(n_obs_events, 48)
    lower_ci = []
    for sample in samples:
        yhat = exp_smoothing_forecast(sample)
        ci = yhat * (1-tolerance_threshold)
        lower_ci.append(ci)
    del lower_ci[-1]
    lower_ci = np.array(lower_ci)
    delta =  np.absolute(len(n_obs_events) - len(lower_ci))
    support_arr = np.full(delta, np.nan)
    lower_ci = np.concatenate((support_arr, lower_ci), axis=None)
    return lower_ci.reshape(-1,1)
    
```

```python
# one-step Holt Winter's Exponential Smoothing forecast
def exp_smoothing_forecast(history):
	# define model model
	history = array(history)
	# run block of code and catch warnings
	with warnings.catch_warnings():
		# ignore all caught warnings
		warnings.filterwarnings("ignore")
		# execute code that will generate warnings
		model = ExponentialSmoothing(history, trend= 'add', damped_trend= False , seasonal='add', seasonal_periods=24)
		# fit model
		model_fit = model.fit(optimized=True, remove_bias = True)
		# make one step forecast
		yhat = model_fit.predict(len(history), len(history))
		
		if yhat[0] < 0:
			yhat[0]  = 0
		else:
			yhat[0] 
	return yhat[0]
```

```python
def is_outlier(obs_n_events, lower_ci):
    # Observed rate estimated from observations
    obs_rate = np.array(([est_poission_rate_ci(obs,1) for obs in obs_n_events]))
    p_outlier = np.array(([rate.cdf(ci) for rate,ci in zip(obs_rate, lower_ci)]))
    yhat = np.array(([ 1 if (p > 0.99) else 0 for p in p_outlier]))
    return yhat.reshape(-1,1) , p_outlier.reshape(-1,1)
```

```python
def condition1(yhat):
    yhat_cd1 = []
    for i in range(len(yhat)):
        yhat_t_24 = yhat[(i-24):i]
        yhat_t = yhat[i]
        if np.sum(yhat_t_24) >= 1 and np.sum(yhat_t) == 1:
            yhat_cd1.append(int(1))
        else:
            yhat_cd1.append(int(0))
    return np.array(yhat_cd1).reshape(-1,1)
```

```python
#partner_id , dt , obs_n_events , lower_ci , p_outlier

def get_outlier_matrix(matrix):
    outlier_matrix = matrix[matrix[:, 6] == 1]
    return outlier_matrix

def send_notification(outlier_matrix):
    for outlier in outlier_matrix:
        outlier = outlier.reshape(-1,1)
        partner_id = outlier[2]
        dt_from = outlier[0]
        dt_to = outlier[1]
        obs_n_events = outlier[3]
        lower_ci = outlier[4]
        p_outlier = outlier[5]
        Pax_lost = obs_n_events - lower_ci
        push_msg_back(partner_id , dt_from, dt_to , obs_n_events , lower_ci , p_outlier)
```

```python
def runner():
    timestamp = ['2022-02-16 10:00:00+00','2022-02-16 11:00:00+00','2022-02-16 12:00:00+00','2022-02-16 13:00:00+00','2022-02-16 14:00:00+00','2022-02-16 15:00:00+00']
    for i in timestamp:
        matrix = get_bigq('marketing_carrier_code','"{}"'.format(i),['FAUT','MARN','ASAL','CEXP','LISC','SAUT','BIGB','REXP','SREG','CAUT','ISAJ','SATA'])
        matrix = matrix.to_numpy()
        matrix = matrix[numpy.argsort(matrix[:, 0])]
        matrix = add_constant(matrix)
        partner_data = extract_partner_data(matrix)
        partner_data_upd = []
        outlier_partner_data = []
        for partner in partner_data:
            lower_ci = get_lower_ci(partner[:,3], 0.2)
            yhat , p_outlier = is_outlier(partner[:,3],lower_ci)
            #yhat_cd1 = condition1(yhat) 
            partner = np.concatenate([partner, lower_ci, p_outlier, yhat], axis = 1)
            partner = sub_constant(partner)
            partner_data_upd.append(partner)
            
        for i in partner_data_upd:
            outlier_matrix = get_outlier_matrix(i)
            outlier_partner_data.append(outlier_matrix)
            send_notification(outlier_matrix)

        print("messages send")
        return partner_data_upd , outlier_partner_data
```
